### Condition Based Filtering

```python
# titanic_df.Fare
mask = titanic_df["Fare"] >= 100
fare_100 = titanic_df[mask]
```

```python
fare_100.shape
```

```python
fare_100.head()
```

```python
female_100 = titanic_df[
    (titanic_df['Fare'] > 100) & (titanic_df['Sex'] == 'female')
]
```
