### Handling NaN or Null Value

```python
titanic_df.isna().sum()
```

1. Remove Null Values
2. Imputation
3. Feature Engineering

```python
# Removing Null Columns
titanic_df[titanic_df['Embarked'].isna()]
```

```python
titanic_df_filter = titanic_df[titanic_df['Embarked'].notna()]

# > dropna
titanic_df.dropna(axis = 0, how = 'any', subset= "Embarked", ignore_index= True)
```

```python
titanic_df_filter.isna().sum()
```

```python
#Imputation (Age) ## fill(Age) value
print(titanic_df_filter.Age.mean())
titanic_df_filter.Age.plot.hist(bins = 20, grid =  True)
```

```python
mean_age = titanic_df['Age'].mean()
mean_age
```

```python
titanic_df_filter['Age'].fillna(value = mean_age, inplace= True)
```

```python
titanic_df_filter.Age.plot.hist(bins = 20, grid =  True)
```

```python
# Assigment: Imputation of Age Column w.r.t gender provided.
titanic_df_filter['Age'][titanic_df_filter['Sex'] == 'female'].fillna
```

```python
titanic_df['Cabin']
```

```python
# Feature from Null Value
titanic_df_filter['Cabin'] = titanic_df_filter['Cabin'].notna().astype(int)
titanic_df_filter.loc[:,'Cabin'] = titanic_df_filter['Cabin'].notna().astype(int)
```

### Handling Outliers

```python
titanic_df_filter.Age.plot.box()
```

```python
Q1 = titanic_df_filter.Age.quantile(q= 0.25)
Q3 = titanic_df_filter.Age.quantile(q= 0.75)
IQR = Q3 - Q1
upper_limit = Q3 + 1.5 * IQR
lower_limit = Q1 - 1.5 * IQR

print(upper_limit, ",", lower_limit)
```

```python
titanic_df_filter[(titanic_df_filter.Age > upper_limit) | (titanic_df_filter.Age < lower_limit)]
```
