**Handling incorrect Data**

```python
import numpy as np
# Create a dummy dataset
data = pd.DataFrame(
    {
        'id' : [1, 2, 3, 4, 5],
        'name' : ['John', 'Jane', 'Bob', 'Alice', 'David'],
        'age' : [25, -31, 42, 19, 37],
        'gender' : ['M', 'F', 'M', 'F', 'Z'],
        'income' : [50000, 70000, np.nan, '30000', 60000]
    }
)
print("Dataset before Handing Incorrect values")
print(data)
```

```python
incorrect_age = data[ (data['age'] <= 0) | (data['age'] > 100)]
incorrect_age.index
```

```python
data.loc[incorrect_age.index, "age"] = abs(data.loc[incorrect_age.index,"age"])
```

```python
data
```

```python
abs(data['age'])
```

```python
incorrect_gender = data[-data['gender'].isin(['M', 'F'])]
data.drop(incorrect_gender.index)
```
