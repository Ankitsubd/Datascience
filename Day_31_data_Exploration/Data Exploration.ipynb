```python
# dataframe.groupby(by=...)
group_by_gender = titanic_df[['Sex', 'Fare', 'Age', 'Survived']].groupby(by= 'Sex')
```

```python
group_by_gender.mean()
```

```python
group_by_embarked = titanic_df[['Fare', 'Age', 'Survived', 'Embarked']].groupby(by= 'Embarked')
```

```python
group_by_embarked.sum() 
```

```python
group_by_embarked_count = titanic_df[['Sex','Fare', 'Age', 'Survived', 'Embarked']].groupby(by= 'Embarked')
```

```python
group_by_embarked_count.count()
```

```python
group_by_embarked_mean = titanic_df[['Fare', 'Age', 'Survived', 'Embarked']].groupby(by=['Embarked', 'Survived'])
```

```python
group_by_embarked_mean.mean()
```
