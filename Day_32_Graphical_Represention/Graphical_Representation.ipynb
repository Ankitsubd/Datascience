```python
import pandas as pd

titanic_df = pd.read_csv("Titanic_data.csv")
```

### 1. Bar Graph

```python
titanic_df['Survived'].value_counts().plot(
    kind = 'bar', legend = True, figsize = (4,4), # figsize chance size
    title = "Count of Survived Passangers", grid = "True",
    fontsize = 10, colormap = 'Dark2'

)
```

```python
group_by_gender = titanic_df[["Sex", "Survived"]].groupby(by='Sex')
group_by_gender.value_counts().plot(kind = "bar", colormap = "Dark2")
```

```python
import matplotlib.pyplot as plt
ax = titanic_df[["Embarked", "Age"]].groupby(by='Embarked').mean().plot(kind="bar")
ax.set_xticklabels(["Port A", "Port B", "Port C"], rotation = 45)
plt.show()
```

### 2. Pie Chart

```python
titanic_df[['Embarked', 'Age']].groupby(by='Embarked').mean().plot(
    kind= 'pie', subplots=True, colormap = "Pastel1"
)
```

```python

titanic_df['Survived'].value_counts().plot(
    kind = 'pie', legend = True, figsize = (4,6),
    title = "Count of Survived Passengers", grid = True,
    fontsize = 10, colormap = "Dark2"
)
```

### 3. Line Graph

```python
titanic_df.loc[:10, "Fare"].plot(kind= 'line')
```

### 4. Scatter Plot(x,y)

```python
titanic_df.plot.scatter("Age", "Fare")
```
