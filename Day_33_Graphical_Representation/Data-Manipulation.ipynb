```python
import pandas as pd

titanic_df = pd.read_csv("Titanic_data.csv")
```

### Inserting and Deleting Column

```python
# Dropping a column
# dataframe.drop(<column_name>, axis)
# axis : {0 or 'index', 1 or 'columns'}, default 0
# Whether to drop lables from the index (0 or 'index') or columns (1 or 'columns')
# inplace = True  : inplace change data

titanic_df.drop("PassengerId", axis = 1)
```

```python
titanic_df.drop("PassengerId", axis = 1, inplace = False)
titanic_df.head()
```

```python
titanic_df.drop("PassengerId", axis = 1, inplace = True)
titanic_df.head()
```

```python
titanic_df.drop([0,2,5], axis = 0, inplace = True)
titanic_df.head()
```
